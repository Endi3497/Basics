## Class

- 1. __init__
        - 객체 안에 데이터를 저장하고 싶다면 필요
        - 객체가 생성될 때 자동으로 실행되는 "초기화 함수"

- 2. __len__
        - 이 객체는 길이가 있다라는 것을 표현하고 싶을 때 필요함.
        - 데이터셋 길이를 알려줘야 하므로 사용
        - Data를 shuffle 하고 epoch를 구성하려면 필요함

- 3. __getitem__
        - dataset[i] 처럼 데이터를 꺼내야 할 때 필요하므로 넣음.
        - DataLoader가 batch를 만들 때 필요함.

## Dataset

- Dataset은 list 처럼 행동해야 하며, 이를 위해서는 다음 두개가 필요함.

        __len__
        __getitem__ 

- 이제는 모델에게 데이터를 떠먹여 주는 로봇인 DataLoader를 짜야함.
        * DataLoader는 Dataset의 __len__()과 __getitem()__ 을 반복 호출해서 batch 만듬.

        from torch.utils.data import DataLoader
        ds = Mydataset(data, labels)
        loader = (ds, batch_size=4, shuffle=True)


- batch 가져오기
        for x_batch, y_batch in loader :
                print(x_batch.shape, y_batch.shape)

## nn.Module

- nn.Module은 PyTorch에서 모델을 만드는 설계도임.

- 두가지 반드시 필요한 함수가 존재함.

        1. __init__(self)

            * 모델의 레이어 또는 가중치를 정의하는 곳.
            * 도로의 설계도

            self.fc = nn.Linear(10,5)
            self.relu = nn.ReLU()

        2. forward(self, x)
            
            * 모델이 입력을 받아서 출력까지 계산하는 실제 연산 함수
            * 실제 사람이 걸어다니는 길

            x = self.fc(x)
            x = self.relu(x)
            return x



- ex) 입력 10차원, 출력 1차원 모델

        import torch.nn as nn

        class firstmodel(nn.Module):
            def __init__(self):
                self.fc1 = nn.Linear(10,5)
                self.relu = nn.ReLU()
                self.fc2 = nn.Linear(5, 1)
            
            def forward(self, x):
                x = self.fc1(x)
                x = self.relu(x)
                x = self.fc2(x)
                return x

        model = firstmodel()
        output = model(input_tensor)

- nn.Modlue 을 쓰면 nn.Module 안에 있는 레이어의 가중치는 PyTorch가 자동으로 관리함.
- model.to(device) 하면 모든 layer 파라미터가 자동으로 디바이스로 이동함
- optimizer = torch.optim.Adam(model.parameters()) 이런식으로 optimizer를 쓸 수 있음
