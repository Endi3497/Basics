# 머신러닝 기본 지식

#### 1. 리스트, 튜플, 딕셔너리


- 리스트 : 데이터를 추가하고 뺼 수 있음. 데이터가 계속 바뀌는 경우

        list = []
        list.append("1")
        list.append("2")

- 튜플 : 값과 구조가 변해서느 안되는 것들.

        image_size = (64, 64)
        kernel = (3, 3)
        coords = (x, y)

- 딕셔너리 : 구조를 이름 기반으로 관리. 모델 config, 하이퍼 파라미터, 저장경로, 라벨 매핑 등등

            config = {'lr': 0.001, 'patch_size': 16}


#### 2. 리스트와 ndarray, tensor
- 리스트 : 단순한 컨테이너, 수학 연산 불가능. 

        x = [1, 2, 3]
        print (x * 2)
        ==> [1, 2, 3, 1, 2, 3]

    * 파일 경로, 라벨링 등 인덱스 위주 작업

- NumPy ndarray: 수학 연산을 빠르게 처리하기 위한 배열 구조. 브로드캐스팅 지원

        import numpy as np
        x = np.array([1, 2, 3])
        print (x * 2)
        ==> [2, 4, 6]

    * 이미지 전처리, reshape, 스케일링

- PyTorch Tensor: ndarray 역할 + GPU 계산 + 자동 미분 가능

        import torch 
    
    실수 1, 2, 3으로 이루어진 tensor를 생성하고 기울기를 추적하겠다.

        x = torch.tensor([1., 2., 3.], requires_grad=True)

    텐서의 각 항목에 대해서 2를 곱하고 모두 더하겠다.
    기울기는 y가 하나의 스칼라 값으로 나와야 계산할 수 있기 때문에 .sum()을 한 것.

        y = (x*2).sum()

    y에 대해서 텐서의 각 실수에 대한 기울기를 계산한다.

        y.backward()

    그 기울기를 출력한다

        print(x.grad)

    모델 입력, loss 계산, gpu 연산

#### 3. ndarray, tensor의 shape /reshape /transpose /axis

- shape는 데이터의 구조 그 자체이다.

        import torch
        x = torch.tensor([1, 2, 3], [4, 5, 6])
        print(x.shape)

        ==> torch.size([2,3])

- shape의 각 숫자는 axis를 나타낸다.

        axis 0 -> 길이 2
        axis 1 -> 길이 3

- 이미지 텐서의 shape
        
        [Batch, Channel, Height, Width]

- 64 * 64 흑백 이미지 10장을 텐서로 만들면

        torch.size([10, 1, 64 ,64])

- reshape는 shape만 바꾸는 연산이다.

        x = torch.arrange(6)   ==  tensor([0, 1, 2, 3, 4, 5]) 
        y = x.reshape(2, 3)

        ==> tensor ([0,1,2], [3,4,5])

- transpose, permute는 axis의 순서를 바꾸는 연산

        x = torch.tensor([1,2,3], [4,5,6])
        y = x.t()

        ==>  tensor([1,4]
                    [2,5]
                    [3,6])

        [Batch, Channel, Height, Width]

        y = x.permute(0, 2, 3, 1)

        ==> [Batch, Height, Width, Channel]

        이미지 64 * 64를 4*4 패치로 나누는 것

        x = torch.tensor([1, 1,64 ,64])
        y = x.reshape(1, 256, 16)
